```python
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

# Initialize the geocoder only once
geolocator = Nominatim(
    user_agent="employee_geocoder (asadamir200@gmail.com)", timeout=10         
)

geocode = RateLimiter(
    geolocator.geocode, min_delay_seconds=1, max_retries=3, error_wait_seconds=5
)


def get_coordinates(street, postal_code, city):
    """
    Returns latitude and longitude for an address.

    Returns:
        (latitude, longitude) or (None, None)
    """
    address = f"{street}, {postal_code} {city}, Germany"

    print(address)
    
    try:
        location = geocode(address)

        if location:
            return location.latitude, location.longitude

    except Exception as e:
        print(f"Error for '{address}': {e}")

    return None, None
```

```python
import pandas as pd

df = pd.read_excel('../data/synthetic_employee_dataset.xlsx')
```

```python
# Create new columns
df["Latitude"] = None
df["Longitude"] = None

for index, row in df.iterrows():

    lat, lon = get_coordinates(
        row["Street_Address"],
        row["Postal_Code"],
        row["City"]
    )

    df.at[index, "Latitude"] = lat
    df.at[index, "Longitude"] = lon

    print(f"{index + 1}/{len(df)} completed")
    break
```

```python
from geopy.geocoders import Nominatim

address = """Robert-Koch-Straße 1
22851 Norderstedt
Germany"""

geolocator = Nominatim(user_agent="address_to_coordinates")
location = geolocator.geocode(address)

if location:
    print(f"Latitude: {location.latitude}")
    print(f"Longitude: {location.longitude}")
else:
    print("Address not found.")
```

```python
from geopy.distance import geodesic

# (latitude, longitude)
point1 = (53.550341, 10.000654)
point2 = (53.694512, 9.984321)

distance = geodesic(point1, point2).kilometers

print(f"Distance: {distance:.2f} km")
```

```python
from geopy.distance import geodesic

# Office coordinates (latitude, longitude)
office = (53.551086, 9.993682)

df["Distance_km"] = df.apply(
    lambda row: geodesic(
        office,
        (row["Latitude"], row["Longitude"])
    ).kilometers,
    axis=1
)

df
```
