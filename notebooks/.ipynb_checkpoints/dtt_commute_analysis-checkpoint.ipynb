# Deutschlandticket Commute Attractiveness Analysis
### Johnson & Johnson Medical GmbH — Robert-Koch-Straße 1, 22851 Norderstedt

**Goal.** Estimate how attractive public transport is for employees commuting to the
Norderstedt site, and how likely they'd be to adopt the Deutschlandticket, using a
**synthetic** employee dataset (no real employee data is used or required).

**Notebook structure**
1. Setup & configuration
2. Synthetic employee dataset generation
3. Public transport connection assessment (nearest-station assignment)
4. Door-to-door commute time model
5. Commute-time grouping
6. Deutschlandticket adoption scoring
7. Model evaluation (train/test classifier, metrics, feature importance)
8. Final summary output for the business
9. Optional interactive map

> **Modelling note:** Since real HVV/AKN journey-planner APIs aren't available in this
> environment, commute times are computed from a transparent, documented simulation
> (walking/feeder access + wait + in-vehicle + transfer + last-mile time) built on a
> curated set of ~39 real stations relevant to a Norderstedt commute. All assumptions
> are stated explicitly in the code so they can be swapped for real HVV API data later.

## 2. Synthetic employee dataset generation

We simulate employees living across **Hamburg and the surrounding commuter belt**
(Kreis Segeberg, Kreis Pinneberg, Kreis Stormarn, and Hamburg city districts).

Approach:
- A curated list of ~29 real districts/towns, each with an approximate centroid and a
  **sampling weight** loosely reflecting plausible employee density (more weight close
  to Norderstedt and to well-populated Hamburg districts, less toward small outlying towns).
- Each employee is placed near their district centroid with random spatial jitter
  (std ≈ 1 km, clipped to ±2.4 km) so addresses don't all collapse onto a single point,
  while staying geographically sensible.
- Demographic attributes (age, salary band, environmental attitude, car ownership) are
  drawn from simple, clearly-labelled distributions — again, entirely synthetic, no
  real employee data.

**No real employee data of any kind is used anywhere in this notebook.**

```python
# name, lat, lon, sampling weight
AREAS = [
    ("Norderstedt-Mitte",           53.6975, 10.0111, 11),
    ("Norderstedt-Garstedt",        53.6907, 10.0028, 6),
    ("Norderstedt-Friedrichsgabe",  53.7050, 9.9700, 4),
    ("Langenhorn",                  53.6608, 9.9814, 8),
    ("Fuhlsbüttel",                 53.6367, 9.9928, 5),
    ("Ohlsdorf",                    53.6316, 10.0143, 4),
    ("Alsterdorf",                  53.6103, 10.0122, 5),
    ("Winterhude",                  53.5940, 10.0087, 6),
    ("Barmbek",                     53.5892, 10.0392, 6),
    ("Eppendorf",                   53.5921, 9.9877, 6),
    ("Eimsbüttel",                  53.5786, 9.9636, 7),
    ("Niendorf",                    53.6367, 9.9439, 6),
    ("Schnelsen",                   53.6383, 9.9328, 5),
    ("Eidelstedt",                  53.6019, 9.9147, 5),
    ("Hamburg-Innenstadt",          53.5511, 9.9937, 6),
    ("Altona",                      53.5503, 9.9352, 4),
    ("Wandsbek",                    53.5747, 10.0797, 4),
    ("Quickborn",                   53.7361, 9.9092, 6),
    ("Ellerau",                     53.7325, 9.9613, 3),
    ("Hasloh",                      53.7086, 9.8916, 2),
    ("Henstedt-Ulzburg",            53.7898, 9.9697, 5),
    ("Kaltenkirchen",               53.8386, 9.9614, 3),
    ("Bad Bramstedt",               53.9142, 9.8778, 2),
    ("Pinneberg",                   53.6626, 9.7961, 3),
    ("Halstenbek",                  53.6167, 9.8500, 3),
    ("Barmstedt",                   53.7856, 9.7717, 2),
    ("Ahrensburg",                  53.6742, 10.2378, 3),
    ("Glinde",                      53.5578, 10.2119, 2),
    ("Reinbek",                     53.5364, 10.2481, 2),
]

def generate_employees(n=N_EMPLOYEES, seed=RNG_SEED):
    rng = np.random.default_rng(seed)
    names, lats, lons, weights = zip(*AREAS)
    weights = np.array(weights, dtype=float)
    probs = weights / weights.sum()
    idx = rng.choice(len(AREAS), size=n, p=probs)

    jitter_std_deg = 0.009      # ~1.0 km std
    max_jitter_deg = 0.022      # clip to ~2.4 km so addresses stay plausible
    lat_jitter = np.clip(rng.normal(0, jitter_std_deg, n), -max_jitter_deg, max_jitter_deg)
    lon_jitter = np.clip(rng.normal(0, jitter_std_deg, n), -max_jitter_deg, max_jitter_deg)

    df = pd.DataFrame({
        "employee_id": [f"EMP{i:04d}" for i in range(1, n + 1)],
        "home_area": [names[i] for i in idx],
        "home_lat": np.array([lats[i] for i in idx]) + lat_jitter,
        "home_lon": np.array([lons[i] for i in idx]) + lon_jitter,
        "age": np.clip(rng.normal(38, 10, n), 21, 64).round().astype(int),
        "salary_band": rng.choice(["Lower", "Middle", "Upper"], size=n, p=[0.25, 0.55, 0.20]),
        "environmental_attitude": np.clip(rng.normal(55, 20, n), 0, 100).round(1),
        "has_car": rng.random(n) < 0.55,
    })
    return df

employees = generate_employees()
print(f"Generated {len(employees)} synthetic employees across {employees['home_area'].nunique()} areas.")
employees.head()
```

## 3. Public transport connection assessment

Real employee-level HVV journey-planner queries aren't available offline, so we assess
connectivity using a curated reference table of **~39 real stations/stops** relevant to
a Norderstedt commute (the U1 line backbone through Norderstedt, plus U2/U3, S-Bahn,
AKN, and regional bus feeders across the wider catchment). Coordinates are approximate —
good enough for a distance-based simulation, not survey-grade geocoding.

Each station carries:
- **mode** (U-Bahn / S-Bahn / AKN / Regional Bus — a rough proxy for reliability & speed)
- **avg_speed_kmh** — effective end-to-end speed including stops
- **peak_headway_min** — how often it runs (drives expected wait time)
- **transfers_to_norderstedt** — 0 if the station sits directly on the U1 corridor that
  serves Norderstedt Mitte, 1 otherwise (requires a change of line)

Each employee is assigned their **nearest station** by straight-line (haversine) distance.

```python
def haversine_km(lat1, lon1, lat2, lon2):
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat, dlon = lat2 - lat1, lon2 - lon1
    a = np.sin(dlat / 2) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2) ** 2
    return 2 * 6371.0088 * np.arcsin(np.sqrt(a))

# name, mode, avg_speed_kmh, peak_headway_min, lat, lon, transfers_to_norderstedt
STATIONS = [
    ("Norderstedt Mitte",     "U-Bahn (U1)", 32, 5,  53.6975, 10.0111, 0),
    ("Garstedt",              "U-Bahn (U1)", 32, 5,  53.6907, 10.0028, 0),
    ("Ochsenzoll",            "U-Bahn (U1)", 32, 5,  53.6822, 9.9931,  0),
    ("Kiwittsmoor",           "U-Bahn (U1)", 32, 6,  53.6712, 9.9853,  0),
    ("Langenhorn Nord",       "U-Bahn (U1)", 32, 6,  53.6631, 9.9819,  0),
    ("Fuhlsbüttel Nord",      "U-Bahn (U1)", 32, 6,  53.6497, 9.9819,  0),
    ("Ohlsdorf",              "U-Bahn/S-Bahn hub", 35, 4, 53.6316, 10.0143, 0),
    ("Alsterdorf",            "U-Bahn (U1)", 32, 6,  53.6103, 10.0122, 0),
    ("Sengelmannstraße",      "U-Bahn (U1)", 32, 5,  53.6021, 10.0234, 0),
    ("Kellinghusenstraße",    "U-Bahn (U1/U3)", 32, 4, 53.5872, 9.9853, 0),
    ("Klosterstern",          "U-Bahn (U1)", 32, 4,  53.5761, 9.9878,  0),
    ("Jungfernstieg",         "U-Bahn hub",  32, 3,  53.5528, 9.9931,  0),
    ("Hamburg Hauptbahnhof",  "S/U/Regional hub", 40, 3, 53.5527, 10.0067, 0),
    ("Wandsbek Markt",        "U-Bahn (U1)", 32, 5,  53.5747, 10.0797, 0),
    ("Wandsbek-Gartenstadt",  "U-Bahn (U1)", 32, 6,  53.5814, 10.0673, 0),
    ("Barmbek",               "U-Bahn (U3)/S-Bahn", 35, 5, 53.5892, 10.0392, 1),
    ("Borgweg",                "U-Bahn (U3)", 32, 6, 53.5942, 10.0202, 1),
    ("Hoheluftbrücke",         "U-Bahn (U3)", 32, 5, 53.5811, 9.9784,  1),
    ("Eppendorfer Baum",       "U-Bahn (U3)", 32, 6, 53.5883, 9.9846,  1),
    ("Christuskirche",         "U-Bahn (U2)", 32, 6, 53.5771, 9.9587,  1),
    ("Niendorf Nord",          "U-Bahn (U2)", 32, 6, 53.6367, 9.9439,  1),
    ("Niendorf Markt",         "U-Bahn (U2)", 32, 6, 53.6294, 9.9524,  1),
    ("Schnelsen (Bus)",        "Bus",         20, 15, 53.6383, 9.9328, 1),
    ("Eidelstedt",             "S-Bahn/AKN hub", 40, 10, 53.6019, 9.9147, 1),
    ("Altona",                 "S-Bahn hub",  40, 5,  53.5503, 9.9352,  1),
    ("Sternschanze",           "S-Bahn/U-Bahn (U3)", 35, 5, 53.5631, 9.9639, 1),
    ("Dammtor",                "S-Bahn",      40, 6,  53.5606, 9.9908,  1),
    ("Quickborn",              "AKN (A1)",    45, 20, 53.7361, 9.9092,  1),
    ("Ellerau",                "AKN (A1)",    45, 20, 53.7325, 9.9613,  1),
    ("Hasloh",                 "AKN (A1)",    45, 25, 53.7086, 9.8916,  1),
    ("Henstedt-Ulzburg",       "AKN (A1/A2)", 45, 20, 53.7898, 9.9697,  1),
    ("Kaltenkirchen",          "AKN (A1)",    45, 25, 53.8386, 9.9614,  1),
    ("Pinneberg",              "S-Bahn (S3/S21)", 45, 15, 53.6626, 9.7961, 1),
    ("Halstenbek",             "S-Bahn (S3/S21)", 40, 15, 53.6167, 9.8500, 1),
    ("Poppenbüttel",           "S-Bahn (S1)", 40, 10, 53.6499, 10.0761, 1),
    ("Ahrensburg",             "Regional/S-Bahn (S4)", 50, 15, 53.6742, 10.2378, 1),
    ("Glinde",                 "Bus/Regional", 25, 20, 53.5578, 10.2119, 1),
    ("Reinbek",                "Regional (S21)", 45, 20, 53.5364, 10.2481, 1),
    ("Bad Bramstedt",          "Regional Bus/AKN", 35, 30, 53.9142, 9.8778, 1),
    ("Barmstedt",              "Regional Bus", 30, 30, 53.7856, 9.7717, 1),
]
stations_df = pd.DataFrame(STATIONS, columns=[
    "station", "mode", "avg_speed_kmh", "peak_headway_min", "lat", "lon", "transfers_to_norderstedt"
])
print(f"{len(stations_df)} reference stations/stops loaded.")
stations_df.head()
```

```python
def nearest_station(home_lat, home_lon, stations_df):
    d = haversine_km(home_lat, home_lon, stations_df["lat"].values, stations_df["lon"].values)
    j = int(np.argmin(d))
    row = stations_df.iloc[j]
    return row["station"], row["mode"], row["avg_speed_kmh"], row["peak_headway_min"], row["transfers_to_norderstedt"], d[j]

results = [nearest_station(lat, lon, stations_df) for lat, lon in zip(employees["home_lat"], employees["home_lon"])]
(employees["nearest_station"], employees["station_mode"], employees["station_speed_kmh"],
 employees["headway_min"], employees["transfers"], employees["dist_to_station_km"]) = zip(*results)

employees["distance_to_workplace_km"] = haversine_km(
    employees["home_lat"], employees["home_lon"], WORKPLACE["lat"], WORKPLACE["lon"]
).round(2)

employees[["home_area", "nearest_station", "station_mode", "dist_to_station_km", "distance_to_workplace_km"]].head()
```

**Connection quality score (0–100).** A simple, explainable composite used later for
adoption scoring and the "best/worst connected areas" summary: higher for faster modes
(U-Bahn > S-Bahn > AKN > Regional Bus) and penalised by distance to the nearest station.

```python
def mode_base_score(mode):
    if "U-Bahn" in mode and "AKN" not in mode:
        return 100
    if "S-Bahn" in mode:
        return 80
    if "AKN" in mode:
        return 60
    if "Regional" in mode:
        return 45
    return 35

employees["connectivity_score"] = np.clip(
    employees["station_mode"].map(mode_base_score) - employees["dist_to_station_km"] * 15, 0, 100
).round(1)
employees[["home_area", "station_mode", "dist_to_station_km", "connectivity_score"]].describe(include="all").loc[
    ["mean", "min", "max"], ["dist_to_station_km", "connectivity_score"]
]
```

## 4. Door-to-door commute time calculation

$$\text{door-to-door} = \text{access to station} + \text{wait} + \text{in-vehicle} + \text{transfer penalty} + \text{last-mile to workplace}$$

Assumptions (all explicit and easy to tune):
- **Access to station:** walking at 4.5 km/h up to 1 km; beyond that, a feeder
  bus/bike at 15 km/h (nobody realistically walks 4 km to a train).
- **Wait:** half the peak headway of the assigned line.
- **In-vehicle:** great-circle distance from the station to the workplace, inflated by a
  **1.3× circuity factor** (real rail/road networks aren't straight lines), divided by
  the line's average effective speed.
- **Transfer penalty:** +8 minutes if a line change is needed to reach the Norderstedt
  corridor, 0 if the station sits directly on it.
- **Last-mile:** ~12 minutes (± noise) walking from the Norderstedt-area station to the
  Robert-Koch-Straße site.
- **±6% random noise** applied to the total to reflect normal day-to-day variability.

```python
WALK_SPEED_KMH = 4.5
FEEDER_SPEED_KMH = 15.0
CIRCUITY_FACTOR = 1.3
LAST_MILE_BASE_MIN, LAST_MILE_STD_MIN = 12, 3
TRANSFER_PENALTY_MIN = {0: 0, 1: 8}

def access_time_min(dist_km):
    if dist_km <= 1.0:
        return (dist_km / WALK_SPEED_KMH) * 60
    return (1.0 / WALK_SPEED_KMH) * 60 + ((dist_km - 1.0) / FEEDER_SPEED_KMH) * 60 + 4

rng = np.random.default_rng(RNG_SEED + 1)

wp_dist_by_station = haversine_km(
    WORKPLACE["lat"], WORKPLACE["lon"], stations_df["lat"].values, stations_df["lon"].values
) * CIRCUITY_FACTOR
wp_dist_map = dict(zip(stations_df["station"], wp_dist_by_station))

employees["walk_to_station_min"] = employees["dist_to_station_km"].apply(access_time_min).round(1)
employees["wait_min"] = (employees["headway_min"] / 2).round(1)
employees["in_vehicle_min"] = (
    employees["nearest_station"].map(wp_dist_map) / employees["station_speed_kmh"] * 60
).round(1)
employees["transfer_min"] = employees["transfers"].map(TRANSFER_PENALTY_MIN)
employees["last_mile_min"] = np.clip(rng.normal(LAST_MILE_BASE_MIN, LAST_MILE_STD_MIN, len(employees)), 4, None).round(1)

noise = rng.normal(1.0, 0.06, len(employees))
employees["door_to_door_min"] = (
    (employees["walk_to_station_min"] + employees["wait_min"] + employees["in_vehicle_min"]
     + employees["transfer_min"] + employees["last_mile_min"]) * noise
).round(1)

employees[["employee_id", "home_area", "nearest_station", "walk_to_station_min", "wait_min",
           "in_vehicle_min", "transfer_min", "last_mile_min", "door_to_door_min"]].head()
```

```python
fig, ax = plt.subplots()
ax.hist(employees["door_to_door_min"], bins=30, color="#1f77b4", edgecolor="white")
ax.axvline(employees["door_to_door_min"].median(), color="crimson", linestyle="--",
           label=f"Median = {employees['door_to_door_min'].median():.0f} min")
ax.set_xlabel("Door-to-door commute time (minutes)")
ax.set_ylabel("Number of employees")
ax.set_title("Distribution of simulated public-transport commute times")
ax.legend()
plt.show()

employees["door_to_door_min"].describe().round(1)
```

## 5. Commute-time grouping

Grouping employees into the bands requested in the assessment: **≤30, ≤45, ≤60, >60 minutes**.

```python
bins = [0, 30, 45, 60, np.inf]
labels = ["≤30 min", "31–45 min", "46–60 min", ">60 min"]
employees["commute_band"] = pd.cut(employees["door_to_door_min"], bins=bins, labels=labels)

band_summary = (
    employees["commute_band"].value_counts(normalize=True).reindex(labels) * 100
).round(1)
band_summary = band_summary.rename("share_of_employees_%").to_frame()
band_summary["n_employees"] = employees["commute_band"].value_counts().reindex(labels)
band_summary
```

```python
fig, ax = plt.subplots()
colors = ["#2ca02c", "#8fbf3f", "#e6a817", "#d62728"]
bars = ax.bar(band_summary.index, band_summary["share_of_employees_%"], color=colors)
ax.bar_label(bars, fmt="%.1f%%")
ax.set_ylabel("Share of employees (%)")
ax.set_title("Employees by door-to-door commute-time band")
plt.show()
```

## 6. Deutschlandticket adoption scoring

A transparent, explainable **composite score (0–100)** per employee, combining:

| Component | Weight | Logic |
|---|---|---|
| Time score | 35% | Sigmoid centred at 45 min — comfortably high under ~35 min, drops off past ~60 min |
| Connectivity score | 25% | Mode quality minus distance-to-station penalty (Section 3) |
| Cost-saving appeal | 15% | Longer car commutes → more fuel/parking cost saved by switching |
| Sustainability attitude | 15% | Synthetic environmental-attitude attribute |
| Car ownership | 10% | Existing car owners skew lower (more inertia / sunk cost) |

Employees are then bucketed into **Low / Medium / High** adoption-potential tiers.

We also generate a synthetic **ground-truth adoption outcome** (`adopted`) by treating
the score as a probability, plus noise — this lets us run a proper held-out evaluation
of a classifier in Section 7, as if we were validating against real pilot/survey data.

```python
rng = np.random.default_rng(RNG_SEED + 2)

time_score = 100 / (1 + np.exp((employees["door_to_door_min"] - 45) / 12))
cost_score = np.clip(employees["distance_to_workplace_km"] * 4.5, 0, 100)
env_score = employees["environmental_attitude"].astype(float)

composite = (
    0.35 * time_score
    + 0.25 * employees["connectivity_score"]
    + 0.15 * cost_score
    + 0.15 * env_score
    + 0.10 * (100 - employees["has_car"].astype(int) * 60)
)
composite += rng.normal(0, 4, len(employees))
composite = np.clip(composite, 0, 100)

employees["time_score"] = time_score.round(1)
employees["cost_score"] = cost_score.round(1)
employees["adoption_score"] = composite.round(1)
employees["adoption_tier"] = pd.cut(
    employees["adoption_score"], bins=[-1, 40, 70, 101], labels=["Low", "Medium", "High"]
)

p_adopt = np.clip(employees["adoption_score"] / 100 + rng.normal(0, 0.05, len(employees)), 0.02, 0.98)
employees["adopted"] = (rng.random(len(employees)) < p_adopt).astype(int)

employees[["employee_id", "door_to_door_min", "connectivity_score", "adoption_score", "adoption_tier", "adopted"]].head()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

tier_counts = employees["adoption_tier"].value_counts().reindex(["Low", "Medium", "High"])
axes[0].bar(tier_counts.index, tier_counts.values, color=["#d62728", "#e6a817", "#2ca02c"])
axes[0].set_title("Employees by Deutschlandticket adoption-potential tier")
axes[0].set_ylabel("Number of employees")
for i, v in enumerate(tier_counts.values):
    axes[0].text(i, v + 3, str(v), ha="center")

axes[1].scatter(employees["door_to_door_min"], employees["adoption_score"],
                 c=employees["adoption_score"], cmap="RdYlGn", alpha=0.6, s=15)
axes[1].set_xlabel("Door-to-door commute time (min)")
axes[1].set_ylabel("Adoption score")
axes[1].set_title("Adoption score vs. commute time")
plt.tight_layout()
plt.show()
```

## 7. Model evaluation

To validate that adoption potential is actually **predictable from observable commute
and demographic features** (not just circularly re-deriving the score formula), we:

1. Train/test-split the synthetic `adopted` outcome (75/25, stratified).
2. Fit a **Logistic Regression** (interpretable baseline) and a **Random Forest**
   (captures non-linearities/interactions), using only the *raw* input features —
   not the adoption score itself.
3. Evaluate with accuracy, precision, recall, F1, ROC-AUC, a confusion matrix, and an
   ROC curve.
4. Inspect feature importance to identify the key drivers of adoption.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
                              roc_auc_score, roc_curve, confusion_matrix, ConfusionMatrixDisplay)

FEATURES_NUM = ["door_to_door_min", "dist_to_station_km", "distance_to_workplace_km",
                 "age", "environmental_attitude", "transfers"]
FEATURES_CAT = ["salary_band", "station_mode", "has_car"]

X = employees[FEATURES_NUM + FEATURES_CAT]
y = employees["adopted"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=RNG_SEED, stratify=y
)

preprocess = ColumnTransformer([
    ("num", StandardScaler(), FEATURES_NUM),
    ("cat", OneHotEncoder(handle_unknown="ignore"), FEATURES_CAT),
])

models = {
    "Logistic Regression": LogisticRegression(max_iter=2000, random_state=RNG_SEED),
    "Random Forest": RandomForestClassifier(n_estimators=400, max_depth=6, random_state=RNG_SEED),
}

fitted = {}
eval_rows = []
for name, clf in models.items():
    pipe = Pipeline([("prep", preprocess), ("clf", clf)])
    pipe.fit(X_train, y_train)
    proba = pipe.predict_proba(X_test)[:, 1]
    pred = pipe.predict(X_test)
    fitted[name] = pipe
    eval_rows.append({
        "model": name,
        "accuracy": accuracy_score(y_test, pred),
        "precision": precision_score(y_test, pred),
        "recall": recall_score(y_test, pred),
        "f1": f1_score(y_test, pred),
        "roc_auc": roc_auc_score(y_test, proba),
    })

eval_df = pd.DataFrame(eval_rows).set_index("model").round(3)
eval_df
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# ROC curves
for name, pipe in fitted.items():
    proba = pipe.predict_proba(X_test)[:, 1]
    fpr, tpr, _ = roc_curve(y_test, proba)
    axes[0].plot(fpr, tpr, label=f"{name} (AUC={roc_auc_score(y_test, proba):.2f})")
axes[0].plot([0, 1], [0, 1], "k--", alpha=0.4)
axes[0].set_xlabel("False Positive Rate"); axes[0].set_ylabel("True Positive Rate")
axes[0].set_title("ROC curves — predicting Deutschlandticket adoption")
axes[0].legend()

# Confusion matrix for the best model by AUC
best_name = eval_df["roc_auc"].idxmax()
best_pred = fitted[best_name].predict(X_test)
cm = confusion_matrix(y_test, best_pred)
ConfusionMatrixDisplay(cm, display_labels=["Not adopted", "Adopted"]).plot(ax=axes[1], colorbar=False, cmap="Blues")
axes[1].set_title(f"Confusion matrix — {best_name}")
plt.tight_layout()
plt.show()

print(f"Best model by ROC-AUC: {best_name}")
```

**Feature importance** — what actually drives predicted adoption:

```python
rf_pipe = fitted["Random Forest"]
ohe = rf_pipe.named_steps["prep"].named_transformers_["cat"]
feat_names = FEATURES_NUM + list(ohe.get_feature_names_out(FEATURES_CAT))
importances = rf_pipe.named_steps["clf"].feature_importances_

fi = pd.Series(importances, index=feat_names).sort_values(ascending=True).tail(12)
fig, ax = plt.subplots(figsize=(8, 6))
ax.barh(fi.index, fi.values, color="#4c72b0")
ax.set_title("Top feature importances — Random Forest (predicting adoption)")
ax.set_xlabel("Importance")
plt.tight_layout()
plt.show()
```
